```python
import numpy as np
def GetTimeUnit(Results,j):
    return [i for x,i in enumerate(Results) if x in range(12*j,12*(j+1))]

def GetFP(Grouped,j):
    return np.array([i for x,i in enumerate(Grouped) if x in range(4*j,len(Grouped),3)]).flatten().tolist()

def GetPeriod(Results,j):
    return [i for x,i in enumerate(Results) if x in range(j,len(Results),4)]

def CreateResultDict(Results, PruneZero = False):
    _5T = GetTimeUnit(Results,0)
    _30T = GetTimeUnit(Results,1)
    _1H = GetTimeUnit(Results,2)

    Grouped = [Results[i:i+4] for i in range(0, len(Results), 4)]

    FP1 = GetFP(Grouped,0)
    FP2 = GetFP(Grouped,1)
    FP10 = GetFP(Grouped,2)

    #get each index of alist if i is in some range
    Period1 = [i for x,i in enumerate(Results) if x in range(0,len(Results),4)]
    Period2 = [i for x,i in enumerate(Results) if x in range(1,len(Results),4)]
    Period3 = [i for x,i in enumerate(Results) if x in range(2,len(Results),4)]
    Period4 = [i for x,i in enumerate(Results) if x in range(3,len(Results),4)]

    #prune the zero values
    if PruneZero:
        _5T = [i for i in _5T if i != 0]
        _30T = [i for i in _30T if i != 0]
        _1H = [i for i in _1H if i != 0]

        FP1 = [i for i in FP1 if i != 0]
        FP2 = [i for i in FP2 if i != 0]
        FP10 = [i for i in FP10 if i != 0]

        Period1 = [i for i in Period1 if i != 0]
        Period2 = [i for i in Period2 if i != 0]
        Period3 = [i for i in Period3 if i != 0]
        Period4 = [i for i in Period4 if i != 0]
    result = { "5T": _5T,
            "30T": _30T,
            "1H": _1H,
            "P1": Period1,
            "P2": Period2,
            "P3": Period3,
            "P4": Period4,
            "FP1": FP1,
            "FP2": FP2,
            "FP10": FP10}
    return result
```
